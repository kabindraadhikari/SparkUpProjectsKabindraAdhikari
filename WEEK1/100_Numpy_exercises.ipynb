#### 63. Create an array class that has a name attribute (★★☆)

````python
class NamedArray(np.ndarray):
    def __new__(cls,array, name = "no name"):
        obj = np.asarray(array).view(cls)
        obj.name = name
        return obj
    def __array_finalize__(self, obj):
        if obj is None : return
        self.name = getattr(obj, 'name', "no name")
    
Z = NamedArray(np.arange(10), "range_10")
print (Z.name)
````

#### 75. How to compute averages using a sliding window over an array? (★★★)

````python
def moving_average(a, n=3):
    ret = np.cumsum(a, dtype = float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n -1:] / n
Z = np.arange(20)
print(moving_average(Z, n=3))
````

#### 76. Consider a one-dimensional array Z, build a two-dimensional array whose first row is (Z[0],Z[1],Z[2]) and each subsequent row is  shifted by 1 (last row should be (Z[-3],Z[-2],Z[-1]) (★★★)

````python
from numpy.lib import stride_tricks

def rolling(a,window):
    shape = (a.size - window + 1 , window)
    strides = (a.strides[0], a.strides[0])
    return stride_tricks.as_strided(a, shape=shape, strides = strides )
Z = rolling(np.arange(10),3)
print(Z)
````

#### 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

````python
P0 = np.random.uniform(-10,10,(10,2))
P1 = np.random.uniform(-10,10,(10,2))
p = np.random.uniform(-10,10,(1,2))

def distance_faster(P0,P1,p):
    v = P1- P0 
    v[:,[0,1]] = v[:,[1,0]]
    v[:,1]*=-1 
    norm = np.linalg.norm(v,axis=1)   
    r = P0 - p
    d = np.abs(np.einsum("ij,ij->i",r,v)) / norm 

    return d

print(distance_faster(P0, P1, p))
````

#### 79. Consider 2 sets of points P0,P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P[j]) to each line i (P0[i],P1[i])? (★★★)

````python
P0 = np.random.uniform(-10, 10, (10, 2))
P1 = np.random.uniform(-10, 10, (10, 2))
P = np.random.uniform(-10, 10, (10, 2))

v = P1[:, None, :] - P0[:, None, :]

r = P0[:, None, :] - P[None, :, :]

cross_prod_mag = v[..., 0] * r[..., 1] - v[..., 1] * r[..., 0]

norm_v = np.sqrt(np.sum(v**2, axis=-1))

D = np.abs(cross_prod_mag) / norm_v
print(D)
````

#### 80. Consider an arbitrary array, write a function that extracts a subpart with a fixed shape and centered on a given element (pad with a `fill` value when necessary) (★★★)

````python
def extract_centered_subpart(Z, center, shape, fill_value=0):

    h, w = Z.shape
    sh, sw = shape
    r, c = center

    start_r = r - sh // 2
    end_r = start_r + sh
    start_c = c - sw // 2
    end_c = start_c + sw

    # Calculate padding needed: (pad_before, pad_after)
    pad_r_before = max(0, -start_r)
    pad_r_after = max(0, end_r - h)
    pad_c_before = max(0, -start_c)
    pad_c_after = max(0, end_c - w)

    # Pad the array Z
    Z_padded = np.pad(Z,
                      ((pad_r_before, pad_r_after), (pad_c_before, pad_c_after)),
                      mode='constant',
                      constant_values=fill_value)

    # Recalculate slice coordinates on the padded array
    p_start_r = start_r + pad_r_before
    p_end_r = end_r + pad_r_before
    p_start_c = start_c + pad_c_before
    p_end_c = end_c + pad_c_before

    # Extract the subpart
    return Z_padded[p_start_r:p_end_r, p_start_c:p_end_c]

Z = np.arange(25).reshape(5, 5)
center = (2, 2)
shape = (3, 3)
print("Original Array:")
print(Z)

print("\nCentered 3x3 on (2,2):")
print(extract_centered_subpart(Z, center, shape))

print("\nCentered 5x5 on (0,0) (requires padding):")
center_pad = (0, 0)
shape_pad = (5, 5)
print(extract_centered_subpart(Z, center_pad, shape_pad, fill_value=-1))
````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
class Symetric(np.ndarray):
    def __setitem__(self, index, value):
        i, j = index
        super(Symetric, self).__setitem__((i, j), value)
        super(Symetric, self).__setitem__((j, i), value)

def as_symetric(Z):
    return Z.view(Symetric)

Z = np.zeros((5, 5))
S = as_symetric(Z)
S[1, 2] = 42
print(S)
````

#### 88. How to implement the Game of Life using numpy arrays? (★★★)

````python
def iterate_gol(Z):
    # Count the number of active neighbors using a convolution-like sum
    # np.roll shifts the array to get neighbors
    N = (np.roll(np.roll(Z, 1, 0), 1, 1) +  # South-East
         np.roll(np.roll(Z, 1, 0), -1, 1) + # South-West
         np.roll(np.roll(Z, -1, 0), 1, 1) + # North-East
         np.roll(np.roll(Z, -1, 0), -1, 1) + # North-West
         np.roll(Z, 1, 0) +  # South
         np.roll(Z, -1, 0) + # North
         np.roll(Z, 0, 1) + # East
         np.roll(Z, 0, -1)) # West

    # 1. Any live cell with fewer than two live neighbors dies (underpopulation)
    # 2. Any live cell with two or three live neighbors lives (survival)
    # 3. Any live cell with more than three live neighbors dies (overpopulation)
    # 4. Any dead cell with exactly three live neighbors becomes a live cell (reproduction)

    # Update rules:
    # Rule 2: Live cell (Z=1) with 2 or 3 neighbors (N=2 or N=3) stays alive
    # Rule 4: Dead cell (Z=0) with 3 neighbors (N=3) comes to life
    Z = (N == 3) | (Z & (N == 2))
    return Z

# Example: create a simple glider pattern (period 4)
Z = np.zeros((10, 10), dtype=int)
Z[1, 2] = 1
Z[2, 3] = 1
Z[3, 1:4] = 1

print("Initial state:")
print(Z)

# Simulate one step
Z = iterate_gol(Z)
print("\nState after 1 iteration:")
print(Z)
````

#### 98. Considering a path described by two vectors (X,Y), how to sample it using equidistant samples (★★★)?

````python
X = np.array([0.0, 1.0, 3.0, 6.0])
Y = np.array([0.0, 1.0, 1.0, 4.0])
n_samples = 10

dx = np.diff(Y)
dy = np.diff(X)
segment_lengths = np.sqrt(dx**2 + dy**2)

path_length = np.concatenate(([0.0], np.cumsum(segment_lengths)))
total_length = path_length[-1]

equidistant_points = np.linspace(0, total_length, n_samples)

X_samp = np.interp(equidistant_points, path_length, X)
Y_samp = np.interp(equidistant_points, path_length, Y)

print("X Sampled:", X_samp)
print("Y Sampled:", Y_samp)
````
